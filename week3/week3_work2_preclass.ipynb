```python
# 欠損値があるか確認
df_house_prices_isnull = df_house_prices.isnull().sum()
# 欠損値が1つでもある特徴量を抽出し、欠損値の割合を表示
df_house_prices_isnull[df_house_prices_isnull > 0] / len(df_house_prices) * 100
```

```python
# 欠損値が5つ以上ある特徴量を抽出
missingno_over5 = df_house_prices.isnull().sum()[df_house_prices.isnull().sum() > 5]
# 欠損値が5つ以上ある特徴量を削除
df_house_prices_nullfeaturedropped = df_house_prices.drop(missingno_over5.index, axis=1)
len(df_house_prices_nullfeaturedropped)
```

```python
# 欠損値があるサンプル（行）を削除
df_house_prices_nullvaluedropped = df_house_prices_nullfeaturedropped.dropna(axis=0)
len(df_house_prices_nullvaluedropped)
```

**【問題6】分布の確認**  
- seabornのsns.distplot()を使い目的変数の分布を表示し、「尖度」と「歪度」も算出する。
- 目的変数に対して対数変換を行う。
- 対数変換したものに対して分布を表示し、「尖度」と「歪度」も算出する。  
対数変換の前後での変化を「正規分布」という言葉を含めて説明してください。このように加工することで予測時の精度を上げることができます。

```python
import seaborn as sns
ax = sns.distplot(df_house_prices_nullvaluedropped['SalePrice'])
```

```python
# 尖度を計算
Skewness1 = df_house_prices_nullvaluedropped['SalePrice'].skew()
print("Skewness1 = ", Skewness1)
# 歪度を計算
Kurtosis1 = df_house_prices_nullvaluedropped['SalePrice'].kurt()
print("Kurtosis1 = ", Kurtosis1)
```

```python
import numpy as np
df_log_trans_SalePrice = df_house_prices_nullvaluedropped['SalePrice'].apply(np.log)
ax = sns.distplot(df_log_trans_SalePrice)
```

```python
# 尖度を計算
Skewness2 = df_log_trans_SalePrice.skew()
print("Skewness2 = ", Skewness2)
# 歪度を計算
Kurtosis2 = df_log_trans_SalePrice.kurt()
print("Kurtosis2 = ", Kurtosis2)
```

**【問題7】相関係数の確認**  
特徴量の確認をしていきます。それぞれに対してマークダウンで説明や考察を書いてください。  
- 相関係数行列のヒートマップを作り、相関係数の高い関係を見つける。
- 目的変数との相関の高い特徴量10個を選び、これらだけの相関係数行列ヒートマップを作る。
- 選んだ10個の特徴量が何か表しているかをKaggleのDataDescriptionの記述を参考に日本語でまとめる。
- 選んだ10個の特徴量の中でお互いの相関係数が高い組み合わせを3つ探し出す。  
《補足》  
重回帰分析などで予測する場合、相関が高すぎる特徴量の組み合わせがあると予測に悪影響が出ることがあります。このような場合の簡単な対処方法として、そういった特徴量の組みを探し出して、１つに絞るということがあります。

```python
# 相関係数を算出
corr_mat1 = df_house_prices_nullvaluedropped.corr(method='pearson')

fig = plt.figure(figsize=(20,20))

# 相関係数行列のヒートマップを作り、相関係数の高い関係を見つける
sns.heatmap(corr_mat1,
            vmin=-1.0,
            vmax=1.0,
            center=0,
            annot=True, # True:格子の中に値を表示
            fmt='.1f',
            xticklabels=corr_mat1.columns.values,
            yticklabels=corr_mat1.columns.values
           )
plt.show()
```

0.8以上が相関関係が高いとした場合、  
'OverallQual'と'SalePrice', 'TotalBsmtSF'と'lstFlrFS', 'GrLivArea'と'TotRmsAbvGrd'の相関関係が高いことが分かる。

```python
# 目的変数との相関の高い特徴量10個を選ぶ
abs_corr_mat_SalePrice = np.fabs(corr_mat1['SalePrice']).drop('SalePrice')
corr_mat_SalePrice_top10_index = abs_corr_mat_SalePrice.nlargest(10).index
print(corr_mat_SalePrice_top10_index)

df_house_prices_corr_top10 = df_house_prices_nullvaluedropped[corr_mat_SalePrice_top10_index].corr(method='pearson')
corr_mat2 = df_house_prices_corr_top10.corr(method='pearson')

# これらだけの相関係数行列ヒートマップを作る
fig = plt.figure(figsize=(10,5))

sns.heatmap(corr_mat2,
            vmin=-1.0,
            vmax=1.0,
            center=0,
            annot=True, # True:格子の中に値を表示
            fmt='.2f',
            xticklabels=corr_mat2.columns.values,
            yticklabels=corr_mat2.columns.values
           )
plt.show()
```
