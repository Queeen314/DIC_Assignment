**【問題1】100日目の米粒の数**

100日目には合計で米粒をいくつもらっているのかを計算するコードを作成し、答えを出力してください。また、1日目から100日目までの「その日にもらう米の数」と「その日までに累計で貰う米の数」の推移を折れ線グラフで表現してください。（縦軸が米の数、横軸が日数）

```python
% matplotlib inline
import matplotlib.pyplot as plt

INIT_RICE_NUM = 1
CNT_DAYS = 100

cnt_rice = INIT_RICE_NUM
oneday_rice_num_list = [1]
total_rice_num_list = [1]

for i in range(1, CNT_DAYS):
    cnt_rice *= 2
    oneday_rice_num_list.append(cnt_rice)
    total_rice_num_list.append(sum(oneday_rice_num_list))

plt.xlabel('days')
plt.ylabel('number of rice')
plt.plot(oneday_rice_num_list, label='oneday_rice_num')
plt.plot(total_rice_num_list, label='total_rice_num')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()

print(oneday_rice_num_list[-1])
print(total_rice_num_list[-1])
```

**【問題2】100日目以外の米粒の数**

この逸話は30日だったり81日だったりと日数が曖昧です。どの話にも対応できるように日数を引数とした 関数 を作成しましょう。関数を作成した上で、コースを受講する日数の間でいくつ米粒をもらえるかを答えてください。前の問題同様に折れ線グラフも出力してください。

```python
def calc_rice_numbers(DAYS):
    INIT_RICE_NUM = 1
    cnt_rice = INIT_RICE_NUM
    oneday_rice_num_list = [1]
    total_rice_num_list = [1]

    for i in range(1, DAYS):
        cnt_rice *= 2
        oneday_rice_num_list.append(cnt_rice)
        total_rice_num_list.append(sum(oneday_rice_num_list))
    return oneday_rice_num_list, total_rice_num_list

oneday_rice_num_list, total_rice_num_list = calc_rice_numbers(120)

plt.xlabel('days')
plt.ylabel('number of rice')
plt.plot(oneday_rice_num_list, label='oneday_rice_num')
plt.plot(total_rice_num_list, label='total_rice_num')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```
