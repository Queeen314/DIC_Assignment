**【問題】栗まんじゅうが太陽系を覆う日**  
1個の栗まんじゅうにバイバインを振りかけた場合、太陽系が栗まんじゅうで覆われるのは何分後でしょうか。これを計算するプログラムを作成してください。


栗まんじゅうや太陽系の体積を調査した上で仮定して計算してください。栗まんじゅうの隙間や重力などは考えず、問題を単純化して構いません。

```python
import math

def calc_kurimanju_filled_time(KURIMANJU_VOLUME, VOLUME):
    double_kurimanju_number = 0

    total_kurimanju_volume = KURIMANJU_VOLUME

    total_kurimanju_volume_list = []
    total_kurimanju_volume_list.append(KURIMANJU_VOLUME)

    while total_kurimanju_volume < VOLUME:
        total_kurimanju_volume *= 2
        total_kurimanju_volume_list.append(total_kurimanju_volume)
        double_kurimanju_number += 1
    kurimanju_filled_time = double_kurimanju_number * 5
    return kurimanju_filled_time, total_kurimanju_volume_list

KURIMANJU_VOLUME = 4/3*math.pi*(0.0025**3)
SOLOR_SYSTEM_VOLUME = 4/3*math.pi*(1.5*10**15)**3
TOKYO_DOME_VOLUME = 124*10**4

time1, volume1 = calc_kurimanju_filled_time(KURIMANJU_VOLUME, SOLOR_SYSTEM_VOLUME)
print("太陽系が栗まんじゅうで覆われるのは{}分後です。".format(time1))

import matplotlib.pyplot as plt
%matplotlib inline
xmin, xmax = 0, len(volume1)
plt.title("Doubled_kurimanju_Volume")
plt.xlabel("number of double kurimanju")
plt.ylabel("volume of kurimanju[m^3]")
plt.plot(volume1)
plt.plot([xmin, xmax],[SOLOR_SYSTEM_VOLUME, SOLOR_SYSTEM_VOLUME], "red", linestyle='dashed')
plt.show()
```

```python
time2, volume2 = calc_kurimanju_filled_time(KURIMANJU_VOLUME, TOKYO_DOME_VOLUME)
print("東京ドームが栗まんじゅうで覆われるのは{}分後です。".format(time2))
xmin, xmax = 0, len(volume2)
plt.title("Doubled_kurimanju_Volume")
plt.xlabel("number of double kurimanju")
plt.ylabel("volume of kurimanju[m^3]")
plt.plot(volume2)
plt.plot([xmin, xmax],[TOKYO_DOME_VOLUME, TOKYO_DOME_VOLUME], "red", linestyle='dashed')
plt.show()
```
