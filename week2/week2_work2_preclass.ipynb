**問題1】乱数の作成**  
平均が(-3, 0)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。  
$% <![CDATA[
\left[
\begin{array}{ccc}
  1.0 & 0.8\\
  0.8 & 1.0
\end{array}
\right] %]]>$

```python
import numpy as np
```

```python
np.random.seed(0)
mean1 = [-3, 0]
conv1 = [[1.0, 0.8], [0.8, 1.0]]
tmp1 = np.random.multivariate_normal(mean1, conv1, 500)
print("len(tmp1) = ", len(tmp1))
```

**【問題2】散布図による可視化**  
問題1で作成したデータ点を散布図により可視化してください。  
散布図はmatplotlibのplt.scatter()を使うことで描けます。

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("scatter")
plt.scatter(tmp1[:, 0], tmp1[:, 1])
plt.show()
```

**【問題3】ヒストグラムによる可視化**  
問題1で作成したデータをヒストグラムにより可視化してください。  
ヒストグラムはplt.hist()を使うことで描けます。

```python
plt.xlabel("x1")
plt.ylabel("fluency")
plt.title("histgram of x1")
plt.hist(tmp1[:, 0], 50)
plt.xlim(-6, 3)
plt.ylim(0, 35)
plt.show()
```

```python
plt.xlabel("x2")
plt.ylabel("fluency")
plt.title("histgram of x2")
plt.hist(tmp1[:, 1], bins=50)
plt.xlim(-6, 3)
plt.ylim(0, 35)
plt.show()
```

**【問題4】データの追加**  
新たに平均が(0, -3)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。  
$% <![CDATA[
\left[
\begin{array}{ccc}
  1.0 & 0.8\\
  0.8 & 1.0
\end{array}
\right] %]]>$  
そして、問題1、4それぞれのデータをひとつの散布図として可視化してください。凡例として問題1のものは0、問題2のものは1を表示してください。

```python
np.random.seed(0)
mean2 = [0, -3]
conv2 = [[1.0, 0.8], [0.8, 1.0]]
tmp2 = np.random.multivariate_normal(mean2, conv2, 500)
print("len(tmp2) = ", len(tmp2))
```

```python
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("scatter")
plt.scatter(tmp1[:, 0], tmp1[:, 1], label='0')
plt.scatter(tmp2[:, 0], tmp2[:, 1], label='1')
plt.legend(loc='upper right')
plt.show()
```

**【問題5】データの結合**  
データはまとめておいた方が後々扱いやすいです。問題1、4で作成したndarrayを 結合 し、  
(1000, 2)のndarrayとしてください。結合は  np.concatenate()やnp.vstack()を使うことで行えます。

```python
tmp3 = np.concatenate([tmp1, tmp2])
tmp3.shape
```

**【問題6】ラベル付け**  
ひとまとめになった1000個のデータそれぞれに対して、問題1、4どちらで作成したものなのかを示す ラベル 付けを行ってください。問題1のものには0、問題4のものには1を対応させます。  
0と1を含むラベルの列を新たに追加し、(1000, 3)のndarrayを作成してください。  
機械学習に使用するデータセットはこのような形になっていることが多いです。

```python
add_column = [0 for x in range(500)] + [1 for x in range(500)]
tmp4 = np.insert(tmp3, 2, add_column, axis=1)
tmp4.shape
```
